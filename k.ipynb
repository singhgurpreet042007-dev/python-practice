```python
#wap to check through a fxn a no. is armstrong or not 
def armstrongnumber(n):
    temp=n
    total=0
    while n>0:
        digit=n%10
        total=total+(digit**3)
        n=n//10
    if total==temp:
        print("the no. is armstrong")
    else:
        print("the no. is not armstrong")
armstrongnumber(153)
```

```python
#basic recursion function
def show(n):
    if(n==0):
        return
    print(n)
    show(n-1)
show(5)
```
